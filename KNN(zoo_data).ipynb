```python
def plot_zoos(zoos,col1,col2):
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.lmplot(x=col1,y=col2,
              data=zoos,
              hue="animal_name",
              fit_reg=False)
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    plt.title("zoos animal")
    plt.show()
```

```python
plot_zoos(zoos,"animal_name","toothed")
```

```python
plot_zoos(zoos,"animal_name","class_type")
```

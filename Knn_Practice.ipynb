```python
def plot_zoos(zoos,col1,col2):
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.lmplot(x=col1,y=col2,
              data=zoos,
              hue="legs",
              fit_reg=False)
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    plt.title("zoos animal")
    plt.show()
```

```python
plot_zoos(zoos,"legs","class_type")
```

```python
plot_zoos(zoos,"class_type","aquatic")
```

```python
plot_zoos(zoos,"domestic","milk")
```

```python
plot_zoos(zoos,"breathes","aquatic")
```
